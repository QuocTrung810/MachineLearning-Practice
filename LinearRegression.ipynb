Hàm f(x) = θ0 + θ1 * x

Hàm MSE (Mean Square Error): 1/n * Σ(y_true[i] - y_prediction[i]) ** 2

Thay đổi hàm MSE để đạo hàm đơn giản hơn:

=> Loss function = 1/2n * Σ(y_true[i] - y_pred[i]) ** 2

Ta có: 1/2n * Σ(y_true[i] - (θ0 + θ1*x[i])) ** 2

Đạo hàm (u ^ n)' = n * u ^ n-1 * n'

Đạo hàm theo θ0 = 2/2n * Σ(y_true[i] - (θ0 + θ1*x[i])) * (Σ(y_true[i] - (θ0 + θ1*x[i])))'

= 1/n * Σ(y_true[i] - (θ0 + θ1*x[i]))

Đạo hàm theo θ1 = 2/2n * Σ(y_true[i] - (θ0 + θ1*x[i])) * (Σ(y_true[i] - (θ0 + θ1*x[i])))'

= 1/n * -Σ(y_true[i] - (θ0 + θ1*x[i])) * x[i]

```python
def fx(x, w):
  return w[0] + w[1]*x

def loss(x, y, w):
  error = 0
  for i in range(len(x)):
    error += (y[i] - fx(x[i], w))**2
  return error/(2 * len(x))

def derivative(x, y, w):
  d0 = 0
  d1 = 0
  for i in range(len(x)):
    d0 += (y[i] - fx(x[i], w))
    d1 += (y[i] - fx(x[i], w)) * -x[i]
  return d0/len(x), d1/len(x)
```
