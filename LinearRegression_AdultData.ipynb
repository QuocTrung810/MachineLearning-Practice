```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder, normalize
from sklearn.model_selection import train_test_split
```

```python
dataFrame = pd.read_csv('https://raw.githubusercontent.com/QuocTrung810/MachineLearning-Practice/main/data/adult.data', header = None)
dataFrame.head(5)
```

```python
LE = LabelEncoder()
dataFrame[1] = LE.fit_transform(dataFrame[1])
dataFrame[3] = LE.fit_transform(dataFrame[3])
dataFrame[5] = LE.fit_transform(dataFrame[5])
dataFrame[6] = LE.fit_transform(dataFrame[6])
dataFrame[7] = LE.fit_transform(dataFrame[7])
dataFrame[8] = LE.fit_transform(dataFrame[8])
dataFrame[9] = LE.fit_transform(dataFrame[9])
dataFrame[13] = LE.fit_transform(dataFrame[13])
dataFrame[14] = LE.fit_transform(dataFrame[14])
dataFrame.astype('float64')
dataFrame.head(5)
```

```python
normalizated0 = normalize([dataFrame.values[:, 2]])
dataFrame[2] = normalizated0.T
normalizated1 = normalize([dataFrame.values[:, 10]])
dataFrame[10] = normalizated1.T
dataFrame.head(5)
```

```python
df = dataFrame.values
x0 = np.ones((df.shape[0], 1))
data = np.concatenate((x0, df), axis = 1)
x = data[:, :-1]
y = data[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state = 42)
```

```python
def fx(x, w):
  fx = 0
  for i in range(data.shape[1]):
    fx += w[i] * x[i]

def loss(x, y, w):
  error = 0
  for i in range(len(x)):
    error += (y[i] - fx(x[i], w)) ** 2
  return error/(2*len(x))

def derivative(x, y, w):
  d = [0] * data.shape[1]
  for i in range(len(x)):
    for j in range(len(d)):
      d[j] += (y[i] - fx(x[i], w)) * x[i][j]
  for i in range(len(d)):
    d[i] /= len(x)
```
